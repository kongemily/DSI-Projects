### Correlations

#### Pearson's correlation coefficient shows a linear relationship between two variables and works only on numerical categories

```python
def plt_correlation(data, title):
    fig, ax = plt.subplots(figsize=(20, 20))
    ax.set_title(title, fontsize=24)
    labels = data.columns
     # Set up mask to be "True" in the upper triangle.
    mask = np.zeros_like(data,dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True
    
    ax.set_xticklabels(labels, fontsize=15)
    ax.set_yticklabels(labels, fontsize=15)
    cmap = sns.diverging_palette(240, 0, as_cmap=True)
    sns.heatmap(
        data, vmin=-1, vmax=1, center=0, linewidths=.5,annot=True)
   
```

```python
#Plot the Pearsons Correlation heatmap to see the relationship against saleprice
columns = train_continuous.columns
pearson = train_continuous[columns].corr(method="pearson")
plt_correlation(pearson, "Pearson's Correlation")
```

```python
#Plot the Pearsons Correlation heatmap to see the relationship against saleprice
columns = train_discrete.columns
pearson = train_discrete[columns].corr(method="pearson")
plt_correlation(pearson, "Pearson's Correlation")
```

### Create Polynomial Features

```python
X_train
```

```python
#Sets target variable 'saleprice' as y
y = X_train['saleprice']
#Sets feature variables (everything but Sale Price and Id) as X
X_train.drop(['id','saleprice'], axis = 1, inplace=True)

#Generates the full polynomial feature table.  
poly = PolynomialFeatures(include_bias=False)
X_poly = poly.fit_transform(X_train)
X_poly.shape
```

```python
#Adds feature names to all polynomial features
X_poly = pd.DataFrame(X_poly,columns=poly.get_feature_names(X_train.columns))

#Generates list of poly feature correlations
X_poly_corrs = X_poly.corrwith(y)
#Shows features most highly correlated (positively) with sale price, poly features included
X_poly_corrs.sort_values(ascending=False).head(20)
```

```python
features_keep = list(abs(X_poly.corrwith(y)).sort_values(ascending=False)[:-300].index)
X = X_poly[features_keep]
```

## Linear Regression

```python
#Train test split (Random state is set for reproducibility)
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42)

#Standardize data
ss = StandardScaler()
X_train_sc = ss.fit_transform(X_train)
X_test_sc = ss.transform(X_test)
```

```python
#Log transformation of target variable
y_train_log = np.log(y_train)
y_test_log = np.log(y_test)
```

```python
#instantiate Model
model = LinearRegression()
```

```python
# Check CVS
cross_val_score(model, X_train, y_train_log, cv=5).mean()
```

```python
# Fit model to training data
model.fit(X_train, y_train_log);
```

```python
# Generate R2 score for training data
model.score(X_train, y_train_log)
```

```python
# Generate R2 score for test data
model.score(X_test, y_test_log)
```

```python
# Generation predictions from model, train data, will still be in log form
y_train_pred_log = model.predict(X_train)

# Convert to original units
y_train_pred = np.exp(y_train_pred_log)
```

```python
# Calculate MSE for train data
mse = metrics.mean_squared_error(y_train, y_train_pred)
print(f'MSE: {mse}')

# RMSE for train data
print(f'RMSE: {np.sqrt(mse)}')
```

```python
# Generation predictions from model, test data, will still be in log form
y_test_pred_log = model.predict(X_test)

# Convert to original units
y_test_pred = np.exp(y_test_pred_log)
```

```python
# Calculate MSE for test data
mse = metrics.mean_squared_error(y_test, y_test_pred)
print(f'MSE: {mse}')

# RMSE for test data
print(f'RMSE: {np.sqrt(mse)}')
```

The RMSE for the training dats is much smaller than that of the test data which indicates that the model is overfit and the R2 score is indicating the same, with the score being much higher (i.e more accurate) for train data than that of test data

## Ridge Regression

```python
#Specify desired values to try for each hyperparameter
alpha = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
param_grid = dict(alpha=alpha)
                 
grid_rg = GridSearchCV(Ridge(), param_grid=param_grid, scoring='r2', verbose=1, n_jobs=-1)
grid_result = grid_rg.fit(X_train_sc, y_train_log)
print('Best Score: ', grid_result.best_score_)
print('Best Params: ', grid_result.best_params_)
```

By using grid search to find an optimal alpha value, the R2 score has improved using Ridge regression

## Lasso Regression

```python
#Specify desired values to try for each hyperparameter
alpha = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
param_grid = dict(alpha=alpha, max_iter = [5000])
                 
grid_ls = GridSearchCV(Lasso(), param_grid=param_grid, scoring='r2', verbose=1, n_jobs=-1)
grid_result = grid_ls.fit(X_train_sc, y_train_log)
print('Best Score: ', grid_result.best_score_)
print('Best Params: ', grid_result.best_params_)
```

## ElasticNet

```python
# find optimal alpha with grid search
alpha = np.arange(0.00011,10000,1000)
l1_ratio = [-1,-0.5, -0.3, 0,.3,.5,.7,1]
param_grid = dict(alpha=alpha, l1_ratio=l1_ratio)

param_grid = dict(alpha=alpha, l1_ratio=l1_ratio)
grid_en = GridSearchCV(ElasticNet(), param_grid=param_grid, scoring='r2', verbose=1, n_jobs=-1)
grid_result = grid_en.fit(X_train_sc, y_train_log)
print('Best Score: ', grid_result.best_score_)
print('Best Params: ', grid_result.best_params_)
```

```python
#creates table of all feature variables and their model coefficients
coef_table = pd.DataFrame(grid_rg.best_estimator_.coef_, index=X_train.columns, columns=['Coefficients'])
#creates column with coefficient magnitudes
coef_table['Magnitude']=abs(coef_table['Coefficients'])
#sorts features by coefficient magnitude (i.e.; predictive power) and shows top 25 most influential features
coef_table.sort_values('Magnitude',ascending=False).head(25)['Coefficients']
```

```python
#Generates a list of the model's predictions for the test bloc
preds = grid_rg.predict(X_test_sc)

#Plots predicted values against actual values in the train/test split.
plt.figure(figsize=(10,8))
plt.scatter(np.exp(preds),y_test,marker = '+')
plt.xlabel('Predicted values')
plt.ylabel('Actual values')

#Plots x=y line
plt.plot([0,np.max(y_test)],[0,np.max(y_test)], c = 'k');
```

```python
#Initializes model for whole training set
rg_full_trainset = grid_rg.best_estimator_
#Applies StandardScaler normalization to whole training set
X_sc = ss.fit_transform(X)
#Log transformation of sale price for whole training set
y_log = np.log(y)

#Re-fits the model
rg_full_trainset.fit(X_sc,y_log)
#Scores the model (training score only, of course)
rg_full_trainset.score(X_sc,y_log)
```
